```python
#Part II: Analyze non-blinking cells.
#(find traces with stddev not exceeding a limit)
```

```python
nb=Meta('stream2test.avi')#'S1.avi') #Loads file
nb
```

```python
#Create a mask of cells not in covariance image.
nb.sub_covar()
```

```python
nb.adjust_zstack() #remove background fluorescence
```

```python
#Sanity Check 1
display(nb.My_Covar)
display(nb.My_Avg)
display(nb.My_Diff) 
```

```python
#Invert the mask.
d=nb.My_Diff
f=(255-d)
```

```python
#Sanity Check 2 (show the resulting Canny edge image using My_Diff)
nb.canny_edge_fn(f, minVal=125, maxVal=235)
```

```python
#Plot the trace of intensity within filtered contours.
nb.plot_cm(125, 235, array=f, canny=True)
```

```python
#Sanity Check 3 (should be small number because most cells blink)
print(len(nb.filtered_contours))
length=len(nb.filtered_contours)
```

```python
nb.create_std(stddev=9)
print(len(nb.stdd))
```

```python
#Plot trace across whole zstack at all indices of contours.
for i in range(len(nb.stdd)):
    nb.ppeaks(ind=i, mph=85, mpd=15,thresh=0.15, minVal=125, maxVal=235, 
             array=f, stddev=9, calc_widths=False, normed=True)
```

```python
n_norms = np.array(nb.normed_trace)
```

```python
x = np.arange(n_norms.shape[1])
offset = 0
for i in n_norms:
    plt.plot(x, i+offset,'-')
    offset+=3
plt.ylim(0,22)
#plt.ylabel('Intensity')
plt.xlabel('Frame')
fig = plt.gcf()
add_scalebar(plt.axes(),matchx=False,matchy=False, hidex=False, sizey=1, sizex=0, labely='Length 1')
fig.set_size_inches(6,5)
plt.show()
```

```python
nb.total_mean_fluor(nb.stdd, blinky=False, smooth = False)
#Average fluorescence of all nonblinky cells throughout video.
```

```python
nb.total_mean_fluor(nb.stdd, ymax=178, ymin=166, window = 15, blinky=False, smooth=True)
```

```python
#Part III: Merge traces of blinking and non-blinking cells.
```

```python
merged = np.concatenate((blink.blinky_contours, nb.stdd))
merged_norms = np.concatenate((blink_norms, n_norms))
```
