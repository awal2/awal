```python
aps = []
for cnt in q.filtered_contours:
    peri=cv2.arcLength(cnt,True)
    if peri>=10 and peri<=70: #ADDED: filters peris
        aps.append([cv2.contourArea(cnt), cv2.arcLength(cnt,True)])#closed curve specified
aps=np.array(aps) #area, peri
```

```python
pl.scatter(aps.T[0],aps.T[1])
pl.xlabel('Area')
pl.ylabel('Perimeter')
pl.show()
```

```python
#Algorithm 3 borrowed formula in alg3 from 'Sizing Bacteria by Image Analysis'
def alg3(area_peri):
    #area_peri=area_peri.T
    widths = []
    for p,a in area_peri: 
        widths.append([(p-np.sqrt(np.power(p,2)-4*np.pi*a))/np.pi])
    
    widths=np.array(widths)
    lens = []
    for p_ind in range(len(area_peri.T[1])):
        peri=area_peri.T[1,p_ind]
        lens.append((peri/2.0)+widths[p_ind]*(1-np.pi/2))
    return widths, lens
```

```python
w,l = alg3(aps)
```

```python
vols=[]
for ind in range(len(w)):
    vols.append((np.pi/4)*np.power(w[ind],2)*(l[ind]-(w[ind]/3)))
```

```python
vols=np.ndarray.flatten(np.array(vols))
```

```python
um_vols=[]
for v in vols:
    um_vols.append(np.power((np.power(v,1/3.0)*27.01/100),3))
um_vols=np.ndarray.flatten(np.array(um_vols))
```

```python
pl.plot(um_vols)
```

```python
listv = list(um_vols[np.logical_not(np.isnan(um_vols))])
```

```python
import scipy.stats as stats

h = sorted(listv)  #sorted
fit = stats.norm.pdf(h, np.mean(h), np.std(h))  #this is a fitting indeed
pl.plot(h,fit,'-o')
pl.hist(h,normed=True)      #use this to draw histogram of your data
pl.show()      
```

```python
np.median(listv), np.mean(listv), "Number of cells: "+str(len(h))
```
